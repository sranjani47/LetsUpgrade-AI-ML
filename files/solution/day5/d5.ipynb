```python
def check_prime(input_number):
    if input_number > 1:
        for i in range(2, input_number):
            if (input_number % i) == 0:
                res = "No"
                break
        else:
            res = "Yes"
    else:
        res = "No"        
    return res

listVar = []
v = ""
lst1 = list(range(3,100,2))
for i in lst1:
    if len(listVar) < 20:
        v = check_prime(i)
        if(v == "Yes"):
            listVar.append(i)
print(listVar)
```
