```python
F = int(input("Enter temperature to be converted into kelvin and celcius - "))
print("Temperature in Kelvin ( K ) = {:.3f}".format(273.5 + ((F - 32.0) * (5.0/9.0))))
print("Temperature in Celcius ( C ) = {:.3f}".format((F - 32) * 5 / 9))
```
